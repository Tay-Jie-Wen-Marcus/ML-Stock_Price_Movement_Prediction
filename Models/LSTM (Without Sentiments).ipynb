```python
import io
from google.colab import drive
drive.mount("gdrive")

#Import the libraries
import math
import pandas_datareader as web
import numpy as np
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
plt.style.use('fivethirtyeight')

data = pd.read_csv(r'/content/gdrive/My Drive/Colab Notebooks/ML/latest_data.csv')
data.head
```

```python
train_data = (data['Low']+data['High'])/2.0
plt.figure(figsize = (18,9))
plt.plot(range(data.shape[0]),train_data)
plt.xticks(range(0,data.shape[0],60),data['Date'].loc[::60],rotation=45)
plt.xlabel('Date',fontsize=20)
plt.ylabel('Mid Price',fontsize=18)
plt.show()
```

```python
dataset = train_data.values
# training_data_len = math.ceil( len(dataset) *.8)
```

```python
#Scale the all of the data to be values between 0 and 1 
scaler = MinMaxScaler(feature_range=(0, 1)) 
# dataset = dataset.reshape(-1, 1)
# scaled_data = scaler.fit_transform(dataset)
```

```python
data["Midvalues"] = train_data
data
```

```python
from sklearn.model_selection import KFold # import KFold
from tensorflow.python.keras.datasets import mnist
from tensorflow.python.keras.layers.core import Dense, Dropout, Activation, Flatten
from tensorflow.python.keras.layers import LSTM
from tensorflow.python.keras.models import Sequential
from tensorflow.python.keras.utils import np_utils


# 1182 data
kf = KFold(n_splits=5, shuffle=False) # Define the split - into 5 folds 
predictions_list = []
for train, test in kf.split(data):

  train_df = data.iloc[train]
  train_reshaped = train_df.Midvalues.values.reshape(-1, 1)
  # display(train_reshaped)
  scaled_train = scaler.fit_transform(train_reshaped)
 
  test_df = data.iloc[test]
  test_reshaped = test_df.Midvalues.values.reshape(-1, 1)
  scaled_test = scaler.transform(test_reshaped)

  # print(len(test_df))
  # print(len(scaled_test))

  #Split the data into x_train and y_train data sets
  X_train=[]
  Y_train = []
  for i in range(60,len(scaled_train)):
      X_train.append(scaled_train[i-60:i])
      Y_train.append(scaled_train[i])

  X_test=[]
  Y_test = []
  y_test_index = []
  for i in range(60,len(scaled_test)):
      X_test.append(scaled_test[i-60:i])
      Y_test.append(scaled_test[i])
      y_test_index.append(test[i])

  # print(y_test_index)

  X_train, Y_train, X_test, Y_test= np.array(X_train), np.array(Y_train), np.array(X_test), np.array(Y_test)
  X_train = np.reshape(X_train, (X_train.shape[0],X_train.shape[1],1))
  X_test = np.reshape(X_test, (X_test.shape[0],X_test.shape[1],1))

  # print(X_train.shape)
  # print(Y_train.shape)
  # print(X_test.shape)
  # print(Y_test.shape)

  model = Sequential()
  model.add(LSTM(units=200, return_sequences=True,input_shape=(X_train.shape[1],1)))
  model.add(LSTM(units=200, return_sequences=False))
  model.add(Dense(units=25))
  model.add(Dense(units=1))
  model.compile(optimizer='adam', loss='mean_squared_error')

  #Train the model
  model.fit(X_train, Y_train, batch_size=10, epochs=10)

  predictions = model.predict(X_test)
  trainPredict_dataset_like = np.zeros(shape=(len(predictions), 1182))
  # put the predicted values in the right field
  trainPredict_dataset_like[:,0] = predictions[:,0]
  # inverse transform and then select the right field
  predictions = scaler.inverse_transform(trainPredict_dataset_like)[:,0]

  # print(predictions)

  rmse=np.sqrt(np.mean(((predictions - Y_test)**2)))
  print(rmse)

  #Plot/Create the data for the graph
  # print(len(predictions))
  predictions = pd.Series(predictions, index = y_test_index, name="Predictions")
  # display(predictions)
  test_df = pd.concat([test_df, predictions], axis=1, sort=False)
  # display(test_df)
  # test_df['Predictions'] = predictions

  predictions_list.extend([0]*60 + list(predictions))

  #Visualize the data
  plt.figure(figsize=(16,8))
  plt.title('Model')
  plt.xlabel('Date', fontsize=18)
  plt.ylabel('Mid Price USD ($)', fontsize=18)
  plt.plot((data['Low']+data['High'])/2.0)
  plt.plot((test_df['Low']+test_df['High'])/2.0)
  plt.plot(test_df['Predictions'])
  plt.legend(['Train', 'Actual', 'Predictions'], loc='lower right')
  plt.xticks(range(0,data.shape[0],60),data['Date'].loc[::60],rotation=45)
  plt.show()
```

```python
last_sect = data

last_sect['Predictions'] = predictions_list
last_sect = last_sect[last_sect['Predictions'] != 0]
def output_direction(df):
  if df['Predictions'] > df['Adj Close']:
    return 0
  else:
    return 1

last_sect['Direction_predicted'] = last_sect.apply(output_direction, axis = 1)
count = 0
for index, row in last_sect.iterrows():
  if row['Direction_predicted'] == row['Direction']:
    count += 1

print(count/len(last_sect))
  
```
