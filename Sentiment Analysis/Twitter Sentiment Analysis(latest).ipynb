```python
#-- Useful libraries for NLTK
import re, string
import pandas as pd
import numpy as np
import nltk
from nltk.tag import pos_tag
from nltk.stem.wordnet import WordNetLemmatizer
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem.porter import *
from nltk import FreqDist
from textblob import TextBlob
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
import random
from nltk import classify
from nltk import NaiveBayesClassifier
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#-- Data (Tweets)
df = pd.read_csv('tweets.csv')
tweets = df.Tweet.tolist()
dates = df.TweetDate.tolist()
```

```python
#-- Removing some noise
def remove_pattern(input_txt, pattern):
    r = re.findall(pattern, input_txt)
    for i in r:
        input_txt = re.sub(i, '', input_txt)        
    return input_txt
def clean_tweets(lst):
    # remove twitter Return handles (RT @xxx:)
    lst = np.vectorize(remove_pattern)(lst, "RT @[\w]*:")
    # remove twitter handles (@xxx)
    lst = np.vectorize(remove_pattern)(lst, "@[\w]*")
    # remove URL links (httpxxx)
    lst = np.vectorize(remove_pattern)(lst, "https?://[A-Za-z0-9./]*")
    # remove special characters, numbers, punctuations (except for #)
    lst = np.core.defchararray.replace(lst, "[^a-zA-Z#]", " ")
    return lst
```

```python
tweets = clean_tweets(tweets)
```

```python
len(tweets)
```

```python
#--  vadar sentiment
analyser = SentimentIntensityAnalyzer()

#-- df for vadar sentiment; coumpound, neg, neu, pos
df = pd.DataFrame()
for i in tweets:
    temp = pd.Series(analyser.polarity_scores(i))
    df = df.append(temp, ignore_index = True)
```

```python
def formatD(tdate):
    tdate= tdate[:-9]
    print(tdate)
    tdate=tdate[6:]+"-"+tdate[0:2]+'-'+tdate[3:5]
    return tdate
```

```python
df['TweetDate'] = dates
```

```python
df['TweetDate']=df.TweetDate.apply(formatD)
```
